### 집합 자료형(set)

```python
# 강의 자료
s1 = set()
print(s1)

# 값 1개 추가하기(add)
s1.add(2)
print(s1)

# 값 여러 개 추가하기(update)
s1.update([4, 5, 6])
print(s1)

# 특정 값 제거하기(remove)
s1.remove(2)
print(s1)

# 중복을 허용하지 않는다.
s1.add(4)
print(s1)
```

```python
# 생각하기: 별로 피라미드 만들기 - 김수한 조교
n = 3
#    *.  2(3-1) 1
#   ** * 1(3-2) 3
#  *** **0(3-3) 5
  
# 방법1
for i in range(n):
  for j in reversed(range(n)):
     if j > i:
       print(' ', end='')
     else:
       print('*', end='')
  for j in range(n):
    if j < i:
      print('*', end='')
  print()

# 방법2
for i in range(1, n+1): 
  print(((n-i)*' ')+((2*i-1)*'*'))
```

## 클래스

```python
class Cal_class:
  def __init__(self, a, b):
    self.a = a
    self.b = b
  def cal_plus(self):
    return self.a + self.b
  def cal_multiply(self):
    return self.a * self.b
  def cal_minus(self):
    return self.a - self.b
  def cal_divide(self):
    return self.a / self.b
```

```python
a = 15
b = 5
cal = Cal_class(a,b)
print(cal.cal_plus())
```

```python
fileDir = 'drive/My Drive/hyeonah/특강(겨울방학,조선대)/Data/YogiyoReviews.txt'
fr = open(fileDir, 'r')
contents= fr.readlines()
fr.close()

for content in contents:
  print(content)
```

```python
reviewList = []
import re
for content in contents:
  contentSplit = content.split(" ## ")
  restaurantName = contentSplit[0]
  content = content.replace(restaurantName,"")

  content = content.strip()
  content = re.sub(" ## "," ",content)
  content = re.sub('[^가-힣]', ' ', content)
  content = re.sub('[\s]+', ' ', content)
  reviewList.append(content)
  print(content)

for eachReview in reviewList:
  print(eachReview)
```

```python
#분 초

import time
kkma = Kkma()

# 여러 문장에서 조사만 추출하기

outDir = 'drive/My Drive/hyeonah/특강(겨울방학,조선대)/Data/YogiyoReviews.txt'
f = open(outDir, 'w')

reviewPOSList = []

for eachReview in reviewList:

  try:

    kkmaList = kkma.pos(eachReview)

    eachReviewPOS = ""


    for eachWord in kkmaList:
      eachWord = list(eachWord)
      eachWord = eachWord[0]+"/"+eachWord[1]
      eachReviewPOS = eachReviewPOS + " " + eachWord
      
      eachReviewPOS = eachReviewPOS.strip()

      reviewPOSList.append(eachReviewPOS)
      f.write(eachReviewPOS + "\n")
      
  except:
    print("Error!")

f.close()
```

## 동시등장 단어 행렬

```python
Sentences = ["연어/NNG 초밥/NNG 맛있/VA 어요/EFN",
             "연인/NNG 탕수육/NNG 선택/NNG 구름/NNG 탕수육/NNG 으로/JKM 곱빼기/NNG 추가/NNG 선택/NNG 맛있/VA 게/ECD 잘/MAG 먹/VV 고/ECE 있/VXV 습니다/EFN",
             "연인/NNG 크림/NNG 새우/NNG 만두/NNG 콜라/NNG 사이즈/NNG 선택/NNG 맛있/VA 었/EPT 는데/ECD 크림/NNG 새우/NNG 가/JKS 너무/MAG 딱/MAG 딱하/VA 어서/ECD 많이/MAG 남기/VV 었/EPT 어요/EFN"]

# 단어의 타입

wordType = set()

for sentence in Sentences:
  # print(sentence)
  words = sentence.split(" ")
  # print(words)
  for word in words:
    wordType.add(word)
print(wordType)
print(len(wordType))
```

```python
import pandas as pd

# 0으로 이루어진 단어 행렬 생성 (초기값)
matrix = pd.DataFrame(0.0, index=wordType, columns=wordType)
print(matrix)
```
